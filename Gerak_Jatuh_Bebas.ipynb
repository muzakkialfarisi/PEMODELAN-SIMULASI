```
import matplotlib.pyplot as plt
import math

x, y, x1, y1, vy, vx, t = [], [], [], [], [], [], []

dt = 0.01
v0 = 50
deg = 35
D = 0.0013
g = -9.806
m = 0.15
pi = 22/7
theta = deg*pi/180

vx.append(v0*math.cos(theta))
vy.append(v0*math.sin(theta))

ax = -(D/m)*vx[0]*v0
ay = g - ((D/m)*vy[0]*v0)

T = -2*v0*math.sin(theta)/(ay)

N = int((T/dt) + 1)

t.append(0)
for i in range(N):
  t.append(t[i]+dt)

x.append(0)
y.append(0)
i = 1
while i <= N:
  vy.append(vy[i-1]+(ay*dt))
  y.append(y[i-1]+(vy[i]*dt))
  vx.append(vx[i-1]+(ax*dt))
  x.append(x[i-1]+(vx[i]*dt))
  i += 1

x1.append(0)
y1.append(0)
j = 1
while j <= N:
  x1.append((vx[0]*t[j])+(0.5*ax*t[j]*t[j]))
  y1.append((vy[0]*t[j])+(0.5*ay*t[j]*t[j]))
  j += 1

print(max(y1), max(x1))
print("NON-IDEAL")
plt.plot(x, y) #numerik
plt.plot(x1, y1) #analitik
plt.show()
```
